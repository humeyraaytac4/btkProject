```python
dataFrame["diagnosis"]
```

```python
#x->feature
x=dataFrame.drop("diagnosis",axis=1).values
#y=wx+b
#y->label
y=dataFrame["diagnosis"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

### Normalization-Standardization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### Classification Model

```python
!pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
```

```python
dataFrame.isnull().sum()
```

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu")) #hidden layer
model.add(Dense(units=15,activation="relu")) #hidden layer
model.add(Dense(units=15,activation="relu")) #hidden layer
#relu assigns its value to 0 if the input is negative, and to itself if it is positive.
#Thus, if the input is negative, the neuron is deactivated. We already fit all our inputs between 0 and 1.

model.add(Dense(units=1,activation="sigmoid")) #output layer
#Sigmoid is used because it will take a value between 0 and 1

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700,verbose=0,validation_data=(x_test,y_test))
#I deliberately overestimate the epoch number to see the difference in the measures I take for the overfitting situation.
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss
```

```python
loss.plot()  ## overfitting
```

### Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
newModel=Sequential()

newModel.add(Dense(units=30,activation="relu"))
newModel.add(Dense(units=15,activation="relu"))
newModel.add(Dense(units=15,activation="relu"))
newModel.add(Dense(units=1,activation="sigmoid"))


newModel.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
newModel.fit(x=x_train, y=y_train, epochs=700, verbose=0,validation_data=(x_test,y_test),callbacks=[earlyStopping])
```

```python
loss=pd.DataFrame(newModel.history.history)
```

```python
loss
```

```python
loss.plot()
```

### Dropout

```python
newModel2=Sequential()

newModel2.add(Dense(units=30,activation="relu"))
newModel2.add(Dropout(0.3))

newModel2.add(Dense(units=15,activation="relu"))
newModel2.add(Dropout(0.3))

newModel2.add(Dense(units=15,activation="relu"))
newModel2.add(Dropout(0.3))

newModel2.add(Dense(units=1,activation="sigmoid"))


newModel2.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
newModel2.fit(x=x_train, y=y_train, epochs=700, verbose=0,validation_data=(x_test,y_test),callbacks=[earlyStopping])
```

```python
loss=pd.DataFrame(newModel2.history.history)
```

```python
loss
```

```python
loss.plot()
```
